```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("winequalityN.csv")
```

```python
df.head()
```

```python
x=df.chlorides
y=df.sulphates
plt.scatter(x,y)
plt.title("WINE QUALITY")
plt.xlabel("chlorides")
plt.ylabel("sulphates")
plt.legend()
plt.show()
```

```python

df.hist()
plt.show()
```

```python
df.boxplot(grid=False)
```

```python
plt.bar(x,y,label="BARCHART",color="red")
plt.title("WINE QUALITY")
plt.xlabel("chlorides")
plt.ylabel("sulphates")
plt.legend()
plt.show()
```

```python
plt.plot(x,y,color="magenta")
plt.title("WINE QUALITY")
plt.xlabel("chlorides")
plt.ylabel("sulphates")
plt.legend()
plt.show()
```

```python
plt.stem(x,y)
plt.title("WINE QUALITY")
plt.show()
```
