```
import pandas as pd
import numpy as np
from statsmodels.stats import weightstats as stests
import math
```

```
x_bar = 110
Mu = 50
Sigma = 35
n = 15

Z = (x_bar-Mu)/(Sigma/math.sqrt(n))
print(Z)

if Z < 0.05:
  print("Null Rejected and Alternative Accepted")
else:
  print("Null Accepted and Alternative Rejected")
```

```
df = pd.read_csv("/content/emp.csv")
DF = df.head()
DF
```

```
DF.describe()
```

```
ztest,pvalue = stests.ztest(DF["number_project"],DF["time_spend_company"],value=125)
print(float(pvalue))
print(ztest)
if pvalue < 0.05:
  print("Null Rejected and Alternative Accepted")
else:
  print("Null Accepted and Alternative Rejected")
```

```
ztest,pvalue = stests.ztest(DF["average_montly_hours"], x2=None ,value=130)
print(pvalue)
print(ztest)

if pvalue < 0.05:
  print("Null Rejected and Alternative Accepted")
else:
  print("Null Accepted and Alternative Rejected")
```

```
ztest,pvalue = stests.ztest(DF["average_montly_hours"],value=0)
print(pvalue)
print(ztest)

if pvalue < 0.05:
  print("Null Rejected and Alternative Accepted")
else:
  print("Null Accepted and Alternative Rejected")
```
